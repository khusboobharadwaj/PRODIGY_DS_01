```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_csv_v2_386866/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_386866.csv")
```

```python
df
```

```python
region_counts = df['Region'].value_counts()
bar_width = 0.9
x = range(len(region_counts.index))

plt.bar(x, region_counts.values, width=bar_width)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')

plt.xticks(x, region_counts.index, rotation=45)
plt.tight_layout()
plt.show()
```
